```python
import numpy as np
from numpy import sqrt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
function = lambda x: -sqrt(9- (x ** 2)/25) + 5

x = np.linspace(-15, 15, 20)
plt.plot(x, function(x))
plt.show()
```

```python
def derivative(x):
    return x/(25 * sqrt(9-(x ** 2)/25))
```

```python
def step(x_new, x_previous, precision, l_r):
    x_list, y_list = [x_new], [function(x_new)]
    
    while abs(x_new - x_previous) > precision:
        x_previous = x_new
        d_x = -derivative(x_previous)
        x_new = x_previous + l_r * d_x
        x_list.append(x_new)
        y_list.append(function(x_new))
    
    plot_gradient(x_list, y_list)
```

```python
def plot_gradient(x_list, y_list):        
    plt.subplot(1,2,2)
    plt.scatter(x_list, y_list, c="g")
    plt.plot(x_list, y_list, c="g")
    plt.plot(x, function(x), c="r")
    plt.title("Gradient")
    plt.show()
    
    plt.subplot(1,2,1)
    plt.scatter(x_list, y_list, c="g")
    plt.plot(x_list, y_list, c="g")
    plt.plot(x, function(x), c="r")
    plt.xlim([1.0, 2.1])
    plt.title("Zoomed Gradient")
    plt.show()
```

```python
step(-10, 0, 0.01, 5)
```
